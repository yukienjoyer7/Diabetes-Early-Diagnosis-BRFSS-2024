```python
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_selection import chi2, f_classif
from sklearn.ensemble import RandomForestClassifier

df_numeric = df.select_dtypes(include='number').drop(columns=['RCSGEND1', 'RCSXBRTH']).dropna(subset=['DIABETE4'])
X = df_numeric.drop(columns=['DIABETE4'])
y = df_numeric['DIABETE4']

imputer = SimpleImputer(strategy='median')
X_imputed = pd.DataFrame(imputer.fit_transform(X), columns=X.columns)

scaler = MinMaxScaler()
X_positive = pd.DataFrame(scaler.fit_transform(X_imputed), columns=X.columns)

chi_scores, chi_p_values = chi2(X_positive, y)

anova_scores, anova_p_values = f_classif(X_imputed, y)

rf_model = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=6)
rf_model.fit(X_imputed, y)
rf_importances = rf_model.feature_importances_

hasil_eksperimen = pd.DataFrame({
    'Nama_Fitur': X.columns,
    'Chi2_P_Value': chi_p_values,
    'ANOVA_P_Value': anova_p_values,
    'RF_Importance': rf_importances
})

hasil_eksperimen = hasil_eksperimen.sort_values(by='RF_Importance', ascending=False).reset_index(drop=True)

print(hasil_eksperimen.head(20))

# hasil_eksperimen.to_excel('eksperimen_seleksi_fitur_diabetes.xlsx', index=False)
```
